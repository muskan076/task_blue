```python
import json
import pandas as pd
with open('py_challange_input.json') as f:
    data = json.load(f)
```

```python
aa = data['Cricket']
aa = aa[0]
aa = aa['formats']
```

```python
team_1 = []
team_2 = []
id_1 = []
provider = []
winner = []
link = []
region = []
start_time = []
international = []
draw = []
points = []
```

```python
## all  except name
#id, provider, winner, httplink, team1, team2, starttime
def get_info():
    for question in aa:
        list_1 = question['regions']
        for val in list_1:           
            list_2 = val['matches']
            for val_1 in list_2:
                team_1.append(val_1['team1'])
                team_2.append(val_1['team2'])
                id_1.append(val_1['id'])
                provider.append(val_1['streaming']['liveStream']['provider'])
                winner.append(val_1['results']['winner'])
                points.append(val_1['results']['points'])

                international.append(val_1['isInternational'])
                draw.append(val_1['results']['resultsWithheld'])
                link.append(val_1['httpLink'])
                start_time.append(val_1['startTime'])
```

```python
# for extracting region
region = []
def region_info():
    for question in aa:
        for l in question['regions']:
            for i in range(len(l['matches'])):
                region.append(l['name'])
    return region
        
    
```

```python
ok = [dict(zip(['match_id'], [id_1[i]])) for i in range(len(id_1))]
ok
```

```python
j = 0
for element in ok:
    element['livestream_provider'] = provider[j]
    element['winner_of_the_match'] = winner[j]
    element['match_link'] = link[j]
    element['region_name'] = region[j]
    element['team1'] = team_1[j]
    element['team2'] = team_2[j]
    element['start_time'] = start_time[j]
    j+=1
        
```
